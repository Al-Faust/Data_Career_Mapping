```python
import numpy as np 
import pandas as pd
import os
import re
```

```python
ds_df = pd.read_csv('all_jobs_lat_lon.csv')
```

```python
#make everything lowercase
ds_df['Job_Description'] = ds_df['Job_Description'].str.lower()

#Let's remove all non-word charachters

regex = re.compile('[^a-zA-Z\']')

ds_df['Job_Description'] = ds_df['Job_Description'].apply(lambda x: regex.sub(' ', x))
```

```python
skill_types= {}

skill_types['Python'] = ['python']
skill_types['R'] = [' r ']
skill_types['Excel'] = ['excel']
skill_types['SQL'] = ['sql']
skill_types['NoSQL'] = ['nosql']
skill_types['Java'] = ['java']
skill_types['Hadoop'] = ['hadoop']
skill_types['Spark'] = ['spark']
skill_types['Hive'] = ['hive']
skill_types['Tableau'] = ['tableau']
skill_types['Power Bi'] = ['power bi']
skill_types['JavaScript'] = ['javascript']
skill_types['SAS'] = ['sas']
skill_types['Kubernetes'] = ['kubernetes']
skill_types['AWS'] = ['aws']


for k,v in skill_types.items():
    skill_types[k] = [skill.lower() for skill in skill_types.get(k)]

def refiner(desc):
    desc = desc.split()
    
    two_word = ''
    
    newskills = []
    
    for word in desc:
        two_word = two_word + word 
        for key,value in skill_types.items():
            if((word in value) or (two_word in value)):
                newskills.append(key)
                
        #check for the two worders, like 'power bi'        
        two_word = word
                
    return list(set(newskills))

#Now all we have to do is apply this do our Job Description
ds_df['Refined_Skills'] = ds_df['Job_Description'].apply(refiner)
ds_df['Refined_Skills']
```
